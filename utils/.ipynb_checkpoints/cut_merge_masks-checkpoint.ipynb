```python
import numpy as np
from rasterio.windows import Window
from typing import List
import os
import rasterio
import torch
import matplotlib.pyplot as plt

from typing import List
import cv2

```

```python
from tqdm import tqdm
from typing import List, Optional
```

# Разрезалка на патчи  

```python
n = ['1', '2', '5', '6_1', '6_2', '9_1', '9_2']

output_folder = '/home/user1/xakaton2024/data/cut_data/'
os.makedirs(output_folder, exist_ok=True)
for t in n:
    image_path = f'/home/user1/xakaton2024/data/given_train/images/{t}.tif' 
    mask_path = f'/home/user1/xakaton2024/data/given_train/masks/{t}.tif' 

    split_image(
        image_path=image_path, mask_path=mask_path,
        output_folder=output_folder, tile_size=640,
        overlap=32, image_id=t
        ) 
```

```python
def get_tiles_with_overlap(image_width: int, image_height: int, 
                           tile_size: int, overlap: int) -> List[Window]:

    step_size = tile_size - overlap
    tiles = []
    for y in range(0, image_height, step_size):
        for x in range(0, image_width, step_size):
            window = Window(x, y, tile_size, tile_size)
            # Adjust window if it exceeds the image bounds
            window = window.intersection(Window(0, 0, image_width, image_height))
            tiles.append(window)
    return tiles
    
def save_tile(src_dataset: rasterio.io.DatasetReader, window: Window, 
              output_folder: str, tile_index: int, image_id: int) -> None:

    transform = src_dataset.window_transform(window)
    tile_data = src_dataset.read(window=window)
    
    profile = src_dataset.profile
    profile.update({
        'driver': 'GTiff',
        'height': window.height,
        'width': window.width,
        'transform': transform
    })
    
    output_filename = os.path.join(output_folder, f"tile_{image_id}_{tile_index}.tif")
    with rasterio.open(output_filename, 'w', **profile) as dst:
        dst.write(tile_data)

def split_image(image_path: str, output_folder: str, mask_path: Optional[str] = None, 
                tile_size: int = 512, overlap: int = 128, image_id: int = 0) -> None:

    with rasterio.open(image_path) as src_image:
        image_width = src_image.width
        image_height = src_image.height

        # Create output directories for images and masks (if available)
        images_folder = os.path.join(output_folder, 'images')
        os.makedirs(images_folder, exist_ok=True)

        if mask_path:
            masks_folder = os.path.join(output_folder, 'masks')
            os.makedirs(masks_folder, exist_ok=True)

        # Get list of tiles with overlap
        tiles = get_tiles_with_overlap(image_width, image_height, tile_size, overlap)

        # Save image tiles (and mask tiles if provided)
        if mask_path:
            with rasterio.open(mask_path) as src_mask:
                for idx, window in tqdm(enumerate(tiles)):
                    save_tile(src_image, window, images_folder, idx, image_id)
                    save_tile(src_mask, window, masks_folder, idx, image_id)
        else:
            for idx, window in tqdm(enumerate(tiles)):
                save_tile(src_image, window, images_folder, idx, image_id)
```

#  Собиралка патчей

```python
 
def merge_binary_masks(output_folder: str, image_id: int, original_width: int, original_height: int, tile_size: int, overlap: int) -> np.ndarray:
    
    reconstructed_image = np.zeros((original_height, original_width), dtype=np.uint8)
    
    # Get sorted tile files
    tile_files = sorted([
        os.path.join(output_folder, 'labels', f)
        for f in os.listdir(os.path.join(output_folder, 'labels'))
        if f.startswith(f'tile_{image_id}_')
    ], key=lambda x: int(x.split('_')[-1].split('.')[0]))
    

    num_tiles_x = (original_width + tile_size - 1) // tile_size + 1  # Ceiling division
    num_tiles_y = (original_height + tile_size - 1) // tile_size + 1 # Ceiling division
    

    for idx, tile_file in enumerate(tile_files):
        with rasterio.open(tile_file) as src:
            tile_data = src.read(1)  # Read the first band
    
            # Calculate the position based on the index
            col = idx % num_tiles_x
            row = idx // num_tiles_x
    
            # Calculate offsets
            x_offset = col * (tile_size - overlap)
            y_offset = row * (tile_size - overlap)
    
            # Calculate the end positions
            height_end = min(y_offset + tile_data.shape[0], original_height)
            width_end = min(x_offset + tile_data.shape[1], original_width)
            
            if y_offset < original_height and x_offset < original_width:
            # Adjust the tile data to fit the reconstructed image
                reconstructed_image[y_offset:height_end, x_offset:width_end] = tile_data[:height_end - y_offset, :width_end - x_offset]
    
    return reconstructed_image



# Example usage


reference_image_path = '/home/user1/xakaton2024/data/given_train/masks/6_1.tif'
original_height, original_width, _ = cv2.imread(reference_image_path).shape
output_folder = '/home/user1/xakaton2024/data/cut_data_6_1/'
image_id = '6_1'
tile_size = 640
overlap = 32

reconstructed_mask = merge_binary_masks(output_folder=output_folder, image_id=image_id, original_width=original_width, original_height=original_height, tile_size=tile_size, overlap=overlap)
```

```python
def display_mask(mask_path: str) -> None:
    """
    Display the binary mask using rasterio and matplotlib.

    Parameters:
        mask_path (str): The path to the TIFF file containing the binary mask.
    """
    with rasterio.open(mask_path) as src:
        mask = src.read(1)  # Read the first band (assuming single-band binary mask)

    # Display the mask
    plt.figure(figsize=(10, 10))
    plt.imshow(mask, cmap='gray', vmin=0, vmax=1)  # Use gray colormap for binary mask
    plt.title('Binary Mask')
    plt.axis('off')  # Hide axes
    plt.show()

def save_mask_as_tiff(mask: np.ndarray, output_path: str, reference_image_path: str) -> None:

    with rasterio.open(reference_image_path) as src:
        profile = src.profile
        profile.update({
            'count': 1,  # Number of bands
            'dtype': 'uint8',  # Data type
            'compress': 'lzw',  # Compression method
            'driver': 'GTiff',  # Output format
            'height': mask.shape[0],  # Height of the mask
            'width': mask.shape[1],  # Width of the mask
            'transform': src.transform,  # Use the same transform as the reference image
            'crs': src.crs  # Use the same CRS as the reference image
        })

        # Save the mask as a TIFF file
        with rasterio.open(output_path, 'w', **profile) as dst:
            dst.write(mask, 1)  # Write the mask to the first band
```
